```python
def getvalues(df_t):
    x1 = df_t['time']
    x2 = df_t['type']
    x3 = df_t['clusters']
    y1 , y2 , y3 = [],[],[]
    for val in (zip(x1,x2,x3)):
        y1.append(val[0])
        y2.append(val[1])
        y3.append(val[2])
    return y1 , y2 , y3
```

```python
# method to get data frame from folder 
# here we are taking 9 files as valid data ( label 1 ) and 2 files as (label 0)
# the assumption here is the behaviour of one user will be anomolus for the other 
```

```python
def getTrainDataset(path):
    valid_user_df = pd.DataFrame()
    invalid_user_df = pd.DataFrame()
    cnt=0
    for fl in glob.glob(path+"/*"):
        if(os.path.isfile(fl)):
            if(cnt<9):
                df = pd.read_csv(fl,error_bad_lines=False,engine='python')
                valid_user_df = pd.concat([valid_user_df,df],axis=0)
            else:
                df = pd.read_csv(fl)
                invalid_user_df = pd.concat([invalid_user_df,df],axis=0)
            cnt+=1
        
    return valid_user_df , invalid_user_df
```

```python
def build_graph(state_size,num_classes,batch_size=None):
    
    x=tf.placeholder(tf.int32,[batch_size,None],'input')
    seq_len=tf.placeholder(tf.int32,[batch_size],'seq_len')
    y=tf.placeholder(tf.int32,[batch_size],'output')
    keep_prob=tf.constant(1.0)
    print("initialization done")
    embeddings = tf.get_variable('embedding_matrix',[7000,state_size])
    rnn_inputs  = tf.nn.embedding_lookup(embeddings,x) 
    print("embedding dome")
    
    cell = tf.contrib.rnn.GRUCell(state_size)
    rnn_outputs , final_state = tf.nn.dynamic_rnn(cell,rnn_inputs,seq_len,dtype=tf.float32)
    print("dybanmic rnn")
    
    with tf.variable_scope('softmax'):
        
        W = tf.get_variable('weight',[state_size,num_classes],initializer=tf.constant_initializer(0.0))
        b = tf.get_variable('bias', [num_classes],initializer = tf.constant_initializer(0.0))
    
    #logits = tf.matmul(final_rnn_op, W) + b
    logits = tf.matmul(final_state, W) + b
    preds = tf.nn.softmax(logits,name='poiu')
    correct = tf.equal( tf.cast(tf.argmax(preds,1), tf.int32 ),y )
    accuracy = tf.reduce_mean(tf.cast(correct,tf.float32), name="accuracy")
    print("accuracy cal culated")
    loss = tf.reduce_mean(tf.nn.sparse_softmax_cross_entropy_with_logits(logits=logits,labels=y))
    train_step = tf.train.AdamOptimizer(1e-4).minimize(loss)
    
    print("finally")
    
    saver = tf.train.Saver(tf.global_variables(), max_to_keep=2)

    
    return{
        
        'x': x,
        'seqlen': seq_len,
        'y': y,
        'dropout': keep_prob,
        'loss': loss,
        'ts': train_step,
        'preds': preds,
        'accuracy': accuracy,
        'checkpt':saver
        }
```

```python
class data_iterator():
    def __init__(self, df,axc):
        self.df = df
        self.axc = axc
        self.size = len(self.df)-1
        self.epochs = 0
        self.shuffle()

    def shuffle(self):
        self.cursor = 0

    def next_batch(self, n):
        if self.cursor+n > self.size:
            self.epochs += 1
            self.cursor = 0
        res = self.df[self.cursor:self.cursor+n]
        label = self.axc[self.cursor:self.cursor+n]
        self.cursor += n
        return res , label
```
